```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
```

```python
audioFile = librosa.example("brahms")
```

```python
y, sr = librosa.load(audioFile, sr = None)
```

```python
energy = np.sum(y ** 2)
print(energy)
```

```python
rms_energy = np.sqrt(np.mean( y** 2))
print(rms_energy)
```

```python
zcr = np.sum(librosa.feature.zero_crossing_rate(y))
print(zcr)
```

```python
spectrogram = np.abs(librosa.stft(y))
print(spectrogram.shape)

librosa.display.specshow(librosa.power_to_db(spectrogram))
plt.colorbar()
plt.title("Spectrogram")
```

```python
mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
print(mel_spectrogram.shape)

librosa.display.specshow(librosa.power_to_db(mel_spectrogram))
plt.colorbar()
plt.xlabel("Time")
plt.ylabel("Hz")
plt.title("Mel Spectrogram")
```

```python
mfcc = librosa.feature.mfcc(y=y, sr=sr)
print(mfcc.shape)

librosa.display.specshow(mfcc)
plt.colorbar()
```

```python
spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
print(spectral_centroid.shape)

plt.semilogy(spectral_centroid.T, label="Spectral Centroid")
plt.ylabel("Hz")
plt.title("Spectral Centroid")
plt.legend()
plt.xlim([0, spectral_centroid.shape[-1]])
```

```python
spectral_rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
print(spectral_rolloff.shape)

plt.semilogy(spectral_rolloff.T, label="Spectral Rolloff")
plt.title("Sprectral Rolloff")
plt.ylabel("Hz")
plt.legend()
plt.xlim([0, spectral_rolloff.shape[-1]])
```

```python
spectral_bandwidth = librosa.feature.spectral_bandwidth(y=y, sr=sr)
print(spectral_bandwidth.shape)

plt.semilogy(spectral_bandwidth.T, label="Spectral bandwidth")
plt.title("Spectral Bandwidth")
plt.legend()
plt.ylabel("Hz")
plt.xlim([0, spectral_bandwidth.shape[-1]])
```
