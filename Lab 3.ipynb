# 1. Discrete Cosine Transform (DCT) on 1D Signal:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import dct, idct

# Generate a 1D signal (you can replace this with your own signal)
signal = np.array([1, 2, 3, 4, 5, 6, 7, 8])

# Perform DCT
dct_coefficients = dct(signal, type=2)  # type=2 for the standard DCT-II

# Reconstruct signal from DCT coefficients (optional)
reconstructed_signal = idct(dct_coefficients, type=2)

# Plotting
plt.subplot(2, 1, 1)
plt.plot(signal, marker='o', label='Original Signal')
plt.title('Original Signal')
plt.legend()

plt.subplot(2, 1, 2)
plt.stem(dct_coefficients, markerfmt='ro', basefmt='k-', label='DCT Coefficients')
plt.title('DCT Coefficients')
plt.legend()

plt.tight_layout()
plt.show()
```
