```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1         # Amplitude
f = 5         # Frequency (Hz)
phi = np.pi/4  # Phase

# Time vector
t = np.linspace(0, 1, 1000, endpoint=False)

# Sinusoidal signal
x = A * np.sin(2 * np.pi * f * t + phi)

# Plotting
plt.plot(t, x)
plt.title('Sinusoidal Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
from scipy.io import wavfile
# Load audio file
fs, audio_data = wavfile.read('bubbles.wav')

# Perform FFT
frequencies = np.fft.fftfreq(len(audio_data), 1/fs)
magnitude = np.abs(np.fft.fft(audio_data))

# Plotting
plt.plot(frequencies, magnitude)
plt.title('FFT of Audio Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```

```python
# Load ECG data (replace this with your actual data)
from scipy.datasets import electrocardiogram
ecg_data = electrocardiogram()

sampling_rate = 360

# Perform FFT
ecg_fft = np.fft.fft(ecg_data)
frequencies_ecg = np.fft.fftfreq(len(ecg_data), 1/sampling_rate)

# Plotting
plt.plot(frequencies_ecg, np.abs(ecg_fft))
plt.title('FFT of ECG Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```
