```python
import numpy as np
from IPython.display import Audio
import matplotlib.pyplot as plt
```

```python
fs = 8000
duration = 2.0
f0 = 50
N = int(duration * fs)
n = np.arange(N) / fs
x = np.cos(2 * np.pi * f0 * n)
```

```python
Audio(x,rate=fs)
```

```python
plt.subplot(6,1,1)
plt.plot(n, x)
plt.title("Original Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
def quantize(x, n_bits):
    bins = np.linspace(x.min(), x.max(), num=2**n_bits, endpoint=False)
    return np.digitize(x, bins)
    
```

```python
for i, bits in enumerate([16,8,4,2,1]):
    quantizedAudio = quantize(x, bits)
    print(f"{bits}-bits")
    
    plt.subplot(6,1, i+2)
    plt.plot(n, quantizedAudio, label=f"{bits}-bits")
    plt.xlabel("Time")
    plt.ylabel("Amplitude")
    
    plt.tight_layout()
    plt.show()
```

```python
decay_factor = 0.95
newX = np.exp(-decay_factor * n) * np.cos(2 * np.pi * f0 * n)
```

```python
Audio(newX, rate=fs)
```

```python
plt.subplot(6,1,1)
plt.plot(n, newX)
plt.title("Original Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
for i, bits in enumerate([16,8,4,2,1]):
    quantizedAudio = quantize(newX, bits)
    print(f"{bits}-bits")
    
    plt.subplot(6,1, i+2)
    plt.plot(n, quantizedAudio, label=f"{bits}-bits")
    plt.xlabel("Time")
    plt.ylabel("Amplitude")
    
    plt.tight_layout()
    plt.show()
```
